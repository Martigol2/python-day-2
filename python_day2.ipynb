## Returns

```python
## Return values
 
def calculate_sum (a, b):
    total = a + b
    print(total )
    return total
result = calculate_sum(10, 20)

print(result)
```

```python
# note: the return statement is used to send a value back to the caller of the function. If a function does not have a return statement, it will return None by default.

def hello_say():
    print("Hello")
    return "Hello"
r1 = hello_say()
print(r1) # this will print "Hello" because the function returns a value.
```

```python
def check_even_odd (num):
    if num > 99:
        return "Number is greater than 99"
    
    
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"
    
result1 = check_even_odd(10) 
result2 = check_even_odd(11)
result3 = check_even_odd(100)
 
print(result1) # this will return "Even"
print(result2) # this will return "Odd"
print(result3) # this will return "Number is greater than 99"
```

```python
#iteraration 1 
def calculate_difference (num1, num2):
    return num1 - num2

difference = calculate_difference(10, 5)
difference2 = calculate_difference(20, 10)
difference3 = calculate_difference(100, 50)

# this will return 5
print(calculate_difference(10, 5)) # this will return 5
print(difference) # this will return 5
print(difference2) # this will return 10
print(difference3) # this will return 50
```

```python
#iteraration 2
def calculate_division (num1, num2):
    if num2 == 0:
        return "Cannot divide by zero"
    else:
        return num1 / num2
    
div1 = calculate_division(10, 2)
div2 = calculate_division(20, 0)
div3 = calculate_division(100, 10)
 
print(div1) # this will return 5.0
print(div2) # this will return "Cannot divide by zero"
print(div3) # this will return 10.0
```

```python
#iteraration 3
def calculate_total(num1, num2, num3):
    return sum([num1, num2, num3])

total1 = calculate_total(5, 10, 15)
total2 = calculate_total(5, 15, 25)
total3 = calculate_total(100, 200, 300)

print(total1) # this will return 60
print(total2) # this will return 45
print(total3) # this will return 600    
```

```python
#iteraration 3
def calculate_total(grades):
    return sum(grades)

total1 = calculate_total([5, 10, 15])
total2 = calculate_total([5, 15, 25])
total3 = calculate_total([100, 200, 300])

print(total1) # this will return 60
print(total2) # this will return 45
print(total3) # this will return 600    
```

```python
def calculate_total(lst):
    if lst:
        total = lst[0]
        for num in lst[1:]:
            total += num
        return total
    return "Error: List is empty"
print(calculate_total([5, 10, 15])) # this will return 30
print(calculate_total([5, 15, 25])) # this will return 45
print(calculate_total([])) # this will return 0
```

## Scope: "LEGB" rules
- local (L)
- Enclosing(E)
- Global (G)
- Built-in (B)

```python
z = 10 #global variable (will be accessible anywhere in the code)

def outer():
    y = 5 # Enclosing variable (will be accessible within the function and any nested functions)
     
    def inner():
        x = 2 # Local variable (will be accessible only within this nested function)
        print(x, y, z) # this will return 2 5 10
    inner()
    

outer()

#print(x) # this will return an error because x is a local variable and cannot be accessed outside the function.
```

## Lamda function

```python
def duplicar(numero):
    return numero * 2
print(duplicar(5)) # this will return 10 

lambda numero: numero * 2
print((lambda numero: numero * 2)(5)) # this will return 10
```

```python
# Definición
sumar = lambda a, b: a + b

# Uso
resultado = sumar(5, 10)
print(resultado)  # Output: 15
```

```python
es_par = lambda x: x % 2 == 0

print(es_par(4))  # Output: True
print(es_par(7))  # Output: False
```

```python
# Imagina que tienes una lista de precios y quieres aplicar un 10% de descuento
precios = [100, 200, 300, 400]

# map() aplica la función lambda a cada elemento de la lista
precios_descuento = list(map(lambda p: p * 0.9, precios))

print(precios_descuento)  # Output: [90.0, 180.0, 270.0, 360.0]
```
